```python
import pandas as pd
import datetime
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer, Reshape, MaxPooling2D, Flatten, Dropout, BatchNormalization
```

```python
df=pd.read_csv('kc_house_data.csv')
```

```python
df.head()
```

```python
df['age']=datetime.datetime.now().year-df['yr_built']
```

```python
df['bedrooms']=df['bedrooms']**2
df['bathrooms']=df['bathrooms']**2
df['sqft_living']=df['sqft_living']**2
```

```python
df['zipcode']=df['zipcode'].astype('category')
```

```python
df.head()
```

```python
x=df.drop(['id','date','lat','long','price'],axis=1)
y=df[['price']]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)  # Eğitim verisini ölçekle
x_test = scaler.transform(x_test)  # Test verisini ölçekle
```

```python
model=Sequential()
model.add(Dense(120,activation='relu'))
model.add(Dense(80,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1))
model.compile(loss='mse',optimizer='adam')


```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=40)
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test,tahmin)
```

```python
mean_squared_error(y_test,tahmin)**0.5
```
